# IntentIQ: Model Experimentation
This notebook compares **Logistic Regression** and **Linear SVM** for intent classification on the LeadIntent AI dataset.

```python
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath('..'))

from utils.preprocess import clean_text
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score

# Load dataset
df = pd.read_csv('../data/dataset.csv')
df['text'] = df['text'].apply(clean_text)

# Split data
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['intent'], test_size=0.2, random_state=42, stratify=df['intent'])

# Vectorize
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

## 1. Logistic Regression

```python
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_vec, y_train)
y_pred_lr = lr_model.predict(X_test_vec)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

## 2. Linear SVM

```python
svm_model = LinearSVC()
svm_model.fit(X_train_vec, y_train)
y_pred_svm = svm_model.predict(X_test_vec)

print("Linear SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

## Conclusion
Both models perform well on this small, curated dataset. **Logistic Regression** is chosen for the final production implementation because it provides native probability scores (`predict_proba`), which are essential for our 'confidence score' requirement in the API.
